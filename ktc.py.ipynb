```python
from sklearn.preprocessing import MinMaxScaler
```

```python
db_marks_array = np.array(db_marks)
```

```python
db_marks_p = MinMaxScaler().fit_transform(X=db_marks_array)
```

```python
clust = KMeans(n_clusters=4, n_jobs=-1)
```

```python
pred = clust.fit_predict(db_marks_p)
```
